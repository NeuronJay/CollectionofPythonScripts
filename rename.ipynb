```python
# import drive and files in google colaboratory
from google.colab import drive
from google.colab import files
drive.mount('/content/drive')
data = "/content/drive/My Drive/Photos/"

```

```python
#import os module
import numpy as np
import os 
arr_img=np.array(os.listdir(data))
```

```python
#define a function for renaming the files
def main(): 
  i = 351
  for filename in arr_img: 
    src = data+ filename 
    dst = data+ str(i) + '.jpg'
    os.rename(src, dst) 
    i += 1
```

```python
#call the main function  
if __name__ == '__main__': 
  main()
```
